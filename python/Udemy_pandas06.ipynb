# サンプルデータの作成

```python
p = ['大阪', '京都', '東京'] * 100
print(p)
```

## データのシャッフル

```python
random.seed(1)
random.shuffle(p)
print(p)
```

```python
random.seed(2)
m = ['お好み焼き'] * 80 + ['カレーライス'] * 120 + ['おでん'] * 50 + ['肉うどん'] * 50
random.shuffle(m)
print(m)
```

```python
df = pd.DataFrame()
df['出身地'] = p
df['好物'] = m
df['年齢'] = stats.norm.rvs(loc=35, scale=10, size=300, random_state=3).astype('int32')
df
```

# ピボットテーブル

```python
df2 = df.pivot_table(index='出身地', columns='好物', values='年齢', aggfunc='mean')
df2
```

## データを抽出して平均値を求める方法

```python
df[(df['出身地'] == '大阪') & (df['好物'] == 'おでん')]['年齢'].mean()
```

```python
df3 = df.pivot_table(index='出身地', columns='好物', values='年齢', aggfunc='min')
df3
```

```python
df2 = df.pivot_table(index='出身地', columns='好物', values='年齢', aggfunc='max')
df2
```

```python
df2 = df.pivot_table(index='出身地', columns='好物', values='年齢', aggfunc='count', margins=True)
df2
```

# クロス集計

```python
pd.crosstab(df['出身地'], df['好物'], margins=True)
```

# 折れ線グラフ

```python
import numpy as np
import matplotlib.pyplot as plt
```

## サンプルデータの作成

```python
x = np.arange(-6.3, 6.4, 0.3)
y1 = np.sin(x)# 正弦関数
y2 = np.cos(x)# 余弦関数
# DataFrameにする
df3 = pd.DataFrame()
df3['x'] = x
df3['sin'] = y1
df3['cos'] = y2
df3.head()
```

```python
df3.plot()# あまりよくない例
```

```python
df3.plot(x='x', y= ['sin', 'cos'])
```

```python
# データ点にマーカーを表示する
df3.plot(x='x', y= ['sin', 'cos'], marker='o')# 点のマーカーを追加、小文字のo
```

```python
df3.plot(x='x', y= ['sin', 'cos'], marker='x')
```

```python
# 線種を設定する
df3.plot(x='x', y= ['sin', 'cos'], ls='-.')
```

```python
# 線の太さ
df3.plot(x='x', y= ['sin', 'cos'], lw=4)
```

```python
# titleの表示
df3.plot(x='x', y= ['sin', 'cos'], title='三角関数')
```

```python
# matplotlibの描画手順に沿った形
plt.figure()
df3.plot(x='x', y= ['sin', 'cos'])
plt.title('三角関数')
plt.grid()#グリッド線
plt.xlim(-6,6)# 横軸の表示範囲
plt.ylim(-1.2, 1.2)
plt.show()
```

# 散布図
## サンプルデータの作成

```python
# 二つの山があるデータセット
x1 = stats.norm.rvs(loc=1, scale=1, size=200, random_state=1)
x2 = stats.norm.rvs(loc=8, scale=2, size=800, random_state=2)
x12 = np.append(x1, x2)
```

```python
len(x12)
```

```python
y1 = stats.norm.rvs(loc=1, scale=1, size=200, random_state=3)
y2 = stats.norm.rvs(loc=8, scale=2, size=800, random_state=4)
y12 = np.append(y1, y2)
```

```python
# DataFrameにする
df4 = pd.DataFrame()
df4['x'] = x12
df4['y'] = y12
```

## プロット

```python
df4.hist(bins=20, figsize=(12, 3))
```

```python
# 散布図
df4.plot(kind='scatter', x='x', y='y')
```

```python
df4.plot.scatter(x='x', y='y', alpha=0.4)
```
